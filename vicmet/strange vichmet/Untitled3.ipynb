```python
from math import * #подключаем модуль имеющий нужные нам математические действия
import matplotlib.pyplot as plt #отрисовка графиков
%matplotlib inline  
```

```python
def function(x): #определяем функцию
    return (sin(x) * sqrt(x) + 1)
```

```python
h = pi/7 
```

```python
mas_x = [0.5 * h, 1,5 * h, 2.5 * h, 4.5 * h, 6.5 * h]
```

```python
def fi(x, mas, i):
    pr = 1
    for j in range(len(mas)):
        if (i != j):
            pr *= ((x - mas[j])/(mas[i] - mas[j]))
    return pr
```

```python
def lagrange(x, mas, i):
    return (function(mas[i]) * fi(x, mas, i))
```

```python
def poly_lagrange(x, mas):
    s = 0
    for i in range(len(mas)):
        k = lagrange(x, mas, i)
        s += k
    return s
```

```python
a = 0
b = 2 * pi
n = 50
```

```python
otrezok = [a + (b-a)/n*i for i in range(n+1)]
```

```python
mas_f = [function(x) for x in otrezok]
```

```python
poly_lag_mas = [poly_lagrange(x, mas_x) for x in otrezok]
```

```python
plt.figure (1)
plt.plot(otrezok, mas_f)
plt.xlim(0,2 * pi)
plt.ylim(-10,10)
for i in range(len(mas_x)):
    plt.scatter(mas_x[i],function(mas_x[i]), s = 30,c="Red")
    lag_mas = [lagrange(x, mas_x, i) for x in otrezok]
    plt.plot(otrezok, lag_mas)
```

```python
plt.figure (2)
plt.plot(otrezok, mas_f)
plt.xlim(0,2 * pi)
plt.ylim(-2.5,5)
for i in range(len(mas_x)):
    plt.scatter(mas_x[i],function(mas_x[i]), s = 30,c="Red")
plt.plot(otrezok, poly_lag_mas)
plt.show
```
