```python
from math import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
h = pi / 7
mas_x = [0.5 * h, 2.5 * h, 6.5 * h, 4.5 * h, 1.5 * h] #данные, которые нам дали
```

```python
def func(x): #оригинальная функция
    return sin(x) * sqrt(x) + 1
```

```python
def neuton(raz_x): #вычисления ньютона
    k = 2
    km = [1 for i in range(len(raz_x))]
    y = []
    for x in raz_x:
        y.append(func(mas_x[0]))
    yield y
    for schet in range(len(mas_x) - 1):
        for per in range(len(raz_x)):
            km[per] *= (raz_x[per] - mas_x[schet])
            vr = 0
            for i in range(k):
                pr = 1
                for j in range(k):
                    if (j != i):
                        pr *= (mas_x[i] - mas_x[j])
                vr += (func(mas_x[i]) / pr)
            vr *= km[per]
            y[per] += vr      
        yield y
        k += 1
```

```python
a = 0
b = 2 * pi
n = 400
mas = [a + (b-a)/n*i for i in range(n+1)] #равномерное разбиение отрезка
```

```python
mo = 1
for i in neuton(mas): #отрисовка графиков
    plt.plot(mas, i, label = "Neuton-%d" % mo)
    mo += 1
plt.xlim(0,2 * pi)
plt.ylim(-3.5, 5.5)

mas_f = [func(x) for x in mas]
plt.plot(mas,mas_f, label = "Orig func")
plt.legend()
plt.show()
```
